```python
# check the number of missing values in each column
sales_data.isnull().sum()
```

```python
# deleting laptop_ID column, this feature is considered irrelevant to this project
sales_data.drop('Date', axis =1, inplace = True)
sales_data.drop('Age_Group', axis =1, inplace = True)
#sales_data.drop('Sub_Category', axis =1, inplace = True)
sales_data.drop('Unit_Cost', axis =1, inplace = True)
sales_data.drop('Unit_Price', axis =1, inplace = True)
sales_data.drop('State', axis =1, inplace = True)
sales_data.drop('Customer_Gender', axis =1, inplace = True)
sales_data.drop('Customer_Age', axis =1, inplace = True)
sales_data.drop('Day', axis =1, inplace = True)
sales_data.drop('Month', axis =1, inplace = True)
sales_data.drop('Product', axis =1, inplace = True)
#sales_data.drop('Order_Quantity', axis =1, inplace = True)
sales_data
```

```python
# to import the label encoder module
from sklearn import preprocessing

# to create a label encoder object
label_encoder = preprocessing.LabelEncoder()

# applying to the categorical columns
sales_data['Year'] = label_encoder.fit_transform(sales_data['Year'])
sales_data['Country'] = label_encoder.fit_transform(sales_data['Country'])
sales_data['Product_Category'] = label_encoder.fit_transform(sales_data['Product_Category'])
sales_data['Sub_Category'] = label_encoder.fit_transform(sales_data['Sub_Category'])
sales_data['Order_Quantity'] = label_encoder.fit_transform(sales_data['Order_Quantity'])
sales_data['Profit'] = label_encoder.fit_transform(sales_data['Profit'])
sales_data['Cost'] = label_encoder.fit_transform(sales_data['Cost'])
sales_data['Revenue'] = label_encoder.fit_transform(sales_data['Revenue'])
```

```python
sales_data
```

```python
# Creating a correlation matrix
multi_check_matrix = sales_data.corr()
multi_check_matrix
```

```python
# To visualize the correlation matrix in a Heatmap
plt.figure(figsize=(20,10))
colormap = sns.color_palette("Blues")
sns.heatmap(multi_check_matrix, cmap=colormap, vmax=0.9, annot=True, linecolor="white", linewidths=0.02)
```

```python
# The Heatmap reveals a high corelation among these independent varibles, so I used the variance_inflation_factor module to remove them
from statsmodels.stats.outliers_influence import variance_inflation_factor
def compute_vif(dataframe):
    VIF=pd.DataFrame()
    VIF["features"]= sales_data.columns
    VIF["vif_values"] = [variance_inflation_factor(dataframe.values, i) for i in range(dataframe.shape[1])]
    return(VIF)
```

```python
# Applying the defined function to the dataframe
compute_vif(sales_data).sort_values(by = 'vif_values', ascending = False)
```

```python
sales_data.drop('Revenue', axis=1)
```

```python
# Creating a new dataframe for the independent varibles known as X
X = sales_data.drop('Profit', axis =1)
X.head()
```

```python
y = sales_data['Profit']
y.head()
```

```python
# To create the train and test dataset
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 10)
```

```python
# importing the Standard Scaler module
from sklearn.preprocessing import StandardScaler

# Create a scaler object
sc = StandardScaler()

# Applying it to the X_train data
X_train = sc.fit_transform(X_train)
pd.DataFrame(X_train)
```

```python
# Applying it to the X_test data
X_test = sc.fit_transform(X_test)
X_test
```

```python
# importing the linear regression model
from sklearn.linear_model import LinearRegression

# Creating a linear regression object
reg = LinearRegression()

# Model traing
reg.fit(X_train, y_train)
```

```python
# To predict the profit, using the model we just built
y_pred = reg.predict(X_test)
y_pred_table=pd.DataFrame(y_pred)
y_pred_table
                 
```

```python
from sklearn import metrics
import numpy as np  
print("Mean Absolute Error:", metrics.mean_absolute_error(y_test, y_pred))
print("Mean Squared Error:", metrics.mean_squared_error(y_test, y_pred))
print("Root Mean Squared Error:",np.sqrt(metrics.mean_absolute_error(y_test, y_pred)) )
```

```python
# Evaluating using the r^2  method, this is the default metric for evaluating a linear model.
reg.score(X_test, y_test)
```

Linear Regression is a good model as its evaluation is above 0.70
